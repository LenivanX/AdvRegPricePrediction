```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
test = pd.read_csv('./temp_data/final_test_data.csv')
test
```

```python
df = pd.read_csv('./temp_data/final_train_data.csv')
df
```

```python
X = df.drop(['SalePrice'], axis=1).copy()
y = df.SalePrice
```

## Linear Regression

```python
model_lr = LinearRegression()
model_lr.fit(X,y)
```

```python
og_test = pd.read_csv('./data/test.csv')
og_test
```

```python
pred_lr = model_lr.predict(test)
pred_lr = pd.DataFrame(pred_lr)
pred_lr
```

# The Id Column

```python
Id = pd.DataFrame(np.array(range(1461,2920)))
Id
```

```python
ans_lr = pd.concat([Id,pred_lr],axis=1)
ans_lr
```

```python
ans_lr.to_csv('./answers/ans_lr.csv', index=False, header=['Id','SalePrice'])
```

## XGBoost

```python
from xgboost import XGBRegressor
```

```python
model_xgb = XGBRegressor()
```

```python
model_xgb.fit(X,y)
```

```python
pred_xgb = model_xgb.predict(test)
pred_xgb = pd.DataFrame(pred_xgb)
ans_xgb = pd.concat([Id,pred_xgb],axis=1)
```

```python
ans_xgb
```

```python
ans_xgb.to_csv('./answers/ans_xgb.csv', index=False, header=['Id','SalePrice'])
```

### Hyper Parameter tuning

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=2)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {
    'booster':['gbtree','dart'],
    'colsample_bytree':[0.8,0.9,1],
    'gamma':[1,2,3],
    'learning_rate':[0.02,0.03,0.04],
    'subsample':[0.8,0.9,1],
    'reg_lambda':[2,3,4,5]
}
```

```python
gscv_xgb = GridSearchCV(XGBRegressor(),params,cv=5,return_train_score=False)
gscv_xgb.fit(X,y)
```

```python
gscv_xgb.best_params_
```

```python
gscv_xgb.best_score_
```

```python
pred_xgb1 = pd.DataFrame(gscv_xgb.predict(test))
pred_xgb1
```

```python
ans_xgb = pd.concat([Id,pred_xgb1],axis=1)
ans_xgb
```

```python
ans_xgb.to_csv('./answers/ans_xgb.csv', index=False, header=['Id','SalePrice'])
```

# ANN

```python
from tensorflow import keras
```

```python
X.shape
```

```python
ann = keras.Sequential([
    keras.layers.Dense(100,input_shape=(260,), activation='relu'),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dense(1, activation='linear')
])
ann.compile(optimizer='adam',loss='mean_squared_error', metrics=['mean_squared_error'])
ann.fit(X,y,epochs=10000)
```

```python
pred_xgb_ann = pd.DataFrame(ann.predict(test))
pred_xgb_ann
```

```python
ans_xgb_ann = pd.concat([Id,pred_xgb_ann],axis=1)
```

```python
ans_xgb_ann.to_csv('./answers/ans_xgb_ann.csv', index=False, header=['Id','SalePrice'])
```
